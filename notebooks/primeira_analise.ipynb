```python
import pandas as pd
import numpy as np

# Caminho do dataset
file_path = "/kaggle/input/datasets/madhu41289/avazu-ctr-prediction-exp/train/train.csv"
df = pd.read_csv(file_path, nrows=1_000_000)


# DIMENSÃO 

print("Dimensão do dataset (linhas, colunas):", df.shape)

print("Primeiras 5 linhas:")
print(df.head())


# VARIÁVEIS EXISTENTES

print("\nNúmero de variáveis:", df.shape[1])

print("Lista das variáveis:")
print(df.columns.tolist())


# TIPOS DE DADOS

print("Tipos de dados:")
print(df.dtypes)


# VALORES NULOS

print("Valores nulos por coluna:")
print(df.isnull().sum())

print("Total de valores nulos no dataset:", df.isnull().sum().sum())


# DUPLICADOS

duplicados = df.duplicated().sum()

print("\nNúmero de linhas duplicadas:", duplicados)

if duplicados == 0:
    print("Não existem linhas duplicadas.")
else:
    print("Existem linhas duplicadas.")


# ESTATÍSTICAS NUMÉRICAS

print(" ESTATÍSTICAS NUMÉRICAS ")

numeric_cols = df.select_dtypes(include=["int64", "float64"])

if len(numeric_cols.columns) > 0:
    print(numeric_cols.describe())
else:
    print("Não existem variáveis numéricas.")


# ESTATÍSTICAS CATEGÓRICAS

print(" ESTATÍSTICAS CATEGÓRICAS ")

categorical_cols = df.select_dtypes(include=["object"])

if len(categorical_cols.columns) > 0:
    print(categorical_cols.describe())
else:
    print("Não existem variáveis categóricas.")
```
