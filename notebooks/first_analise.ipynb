```python
import pandas as pd
import numpy as np
from IPython.display import display

file_path = "/kaggle/input/datasets/madhu41289/avazu-ctr-prediction-exp/train/train.csv"
chunk_size = 1_000_000

total_rows = 0
null_counts = None
duplicate_count = 0
columns = None
dtypes = None

seen_hashes = set()
sample_data = []

print("A processar dataset em blocos...")

for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunk_size)):
    
    total_rows += len(chunk)
    
    # Mostrar progresso de forma limpa (sem spam)
    if (i + 1) % 5 == 0:
        print(f"{(i+1)} milhões de linhas processadas...")
    
    # Guardar primeiras 5 linhas
    if i == 0:
        first_rows = chunk.head()
        columns = chunk.columns.tolist()
        dtypes = chunk.dtypes
    
    # Guardar amostra até 2M linhas
    if len(sample_data) < 2:
        sample_data.append(chunk)
    
    # Valores nulos
    if null_counts is None:
        null_counts = chunk.isnull().sum()
    else:
        null_counts += chunk.isnull().sum()
    
    # Duplicados reais entre blocos
    row_hashes = pd.util.hash_pandas_object(chunk, index=False)
    for h in row_hashes:
        if h in seen_hashes:
            duplicate_count += 1
        else:
            seen_hashes.add(h)

# Criar dataframe de amostra
df_sample = pd.concat(sample_data, ignore_index=True)

print(" RESULTADOS GERAIS ")

# PRIMEIRAS 5 LINHAS 

print(" PRIMEIRAS 5 LINHAS ")

display (first_rows)

# DIMENSÃO
print("Dimensão do dataset (linhas, colunas):", (total_rows, len(columns)))

print("Número de variáveis:", len(columns))

print("Lista das variáveis:")
display(pd.DataFrame(columns, columns=["Variáveis"]))

print("\nTipos de dados:")
display(dtypes.to_frame(name="Tipo de Dado"))

print("\nValores nulos por coluna:")
display(null_counts.to_frame(name="Valores Nulos"))

print("Total de valores nulos:", int(null_counts.sum()))

print("Número total de duplicados reais:", duplicate_count)



# ESTATÍSTICAS NUMÉRICAS


print(" ESTATÍSTICAS NUMÉRICAS ")

numeric_cols = df_sample.select_dtypes(include=["int64", "float64"])

if len(numeric_cols.columns) > 0:
    display(numeric_cols.describe(percentiles=[0.25, 0.5, 0.75]))
else:
    print("Não existem variáveis numéricas.")



# ESTATÍSTICAS CATEGÓRICAS


print(" ESTATÍSTICAS CATEGÓRICAS ")

categorical_cols = df_sample.select_dtypes(include=["object"])

if len(categorical_cols.columns) > 0:
    display(categorical_cols.describe())
else:
    print("Não existem variáveis categóricas.")
```
